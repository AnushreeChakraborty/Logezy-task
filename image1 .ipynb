```python
import os
from PIL import Image
import colorsys
import numpy as np
```

```python
#contains the path of every image
r = []
path=r"D:\Logezy\challenge"
for files in os.listdir(path):
    if(files.endswith(".JPG")):
        r.append(os.path.join(path, files))
    
    if(files.endswith(".jpg")):
        r.append(os.path.join(path, files))
#print(r)
#len(r)
```

```python
#list of names
list1=[]
for files in os.listdir(path):
    if(files.endswith(".JPG")):
        list1.append(files)
        
    if(files.endswith(".jpg")):
        list1.append(files)
#print(list1)
```

```python
#df
import pandas as pd
n=len(list1)
col_names=["name","height","width","avg lightness","avg saturation","Number of pixels with hue in the [0/12 - 1/12] range",
           "Number of pixels with hue in the [1/12 - 2/12] range","Number of pixels with hue in the [2/12 - 3/12] range",
          "Number of pixels with hue in the [3/12 - 4/12] range","Number of pixels with hue in the [4/12 - 5/12] range",
          "Number of pixels with hue in the [5/12 - 6/12] range","Number of pixels with hue in the [6/12 - 7/12] range",
          "Number of pixels with hue in the [7/12 - 8/12] range","Number of pixels with hue in the [8/12 - 9/12] range",
          "Number of pixels with hue in the [9/12 - 10/12] range","Number of pixels with hue in the [10/12 - 11/12] range",
          "Number of pixels with hue in the [11/12 - 12/12] range"]
data=pd.DataFrame(index=range(n), columns=col_names)
```

```python
#h, w lists of every image
#data["name"]
from PIL import Image
w=[]
h=[]
for i in range(n):
    image = Image.open(r[i])
    width, height = image.size
    w.append(width)
    h.append(height)
#print(w,h)
```

```python
#hls-hue, lightness, saturation

pic=[]
pic_pix=[]
v=[]
hls1=[]
hue=[]
lightness=[]
saturation=[]
img=[]
hue_val=[]
hueeee = np.zeros((n,12))
#print(hueeee)
index = 0

for i in r:
    img= Image.open(i)
    r,g,b=img.split()
    r_pix=np.array(list(r.getdata()))
    g_pix=np.array(list(g.getdata()))
    b_pix=np.array(list(b.getdata()))
    #print(r_pix)
    
    hls1=[]
    #hueeee=np.zeros(12)
    
    for rd, gn, bl in zip(r_pix, g_pix, b_pix):
        hls=colorsys.rgb_to_hls(rd/255., gn/255., bl/255.)   #normalize every pixel
        hls1.append(np.array(hls))
        #h=hls1[0]

        #hls1=np.array(hls1)
        hue = np.array(hls[0])
        #hue=np.array(hue)
        #print(hue)
        #hue=np.array(hue)
        #hue_val=hue_check(hue[0])
        if hue >= 0/12 and hue < 1/12 :
            hueeee[index][0] = hueeee[index][0] + 1
        elif hue >= 1/12 and hue < 2/12:
            hueeee[index][1] = hueeee[index][1] + 1
        elif hue >= 2/12 and hue < 3/12:
            hueeee[index][2] = hueeee[index][2] + 1
        elif hue >= 3/12 and hue < 4/12:
            hueeee[index][3] = hueeee[index][3] + 1
        elif hue >= 4/12 and hue < 5/12:
            hueeee[index][4] = hueeee[index][4] + 1
        elif hue >= 5/12 and hue < 6/12:
            hueeee[index][5] = hueeee[index][5] + 1
        elif hue >= 6/12 and hue < 7/12:
            hueeee[index][6] = hueeee[index][6] + 1
        elif hue >= 7/12 and hue < 8/12:
            hueeee[index][7] = hueeee[index][7] + 1
        elif hue >= 8/12 and hue < 9/12:
            hueeee[index][8] = hueeee[index][8] + 1
        elif hue >= 9/12 and hue < 10/12:
            hueeee[index][9] = hueeee[index][9] + 1
        elif hue >= 10/12 and hue < 11/12:
            hueeee[index][10] = hueeee[index][10] + 1
        elif hue >= 11/12 and hue < 12/12:
            hueeee[index][11] = hueeee[index][11] + 1
           
    #print(hueeee[index][0])
    #print(index)
    index = index + 1   
    hls1=np.array(hls1)
    avg_l_s=hls1.mean(axis=0)  #row mean
    lightness.append(avg_l_s[1])
    saturation.append(avg_l_s[2])

#print(hue)
```

```python
index = 0
for i in range(n):
    #print(n)
    data.iloc[i,0]=list1[i]
    data.iloc[i,1]=h[i]
    data.iloc[i,2]=w[i]
    data.iloc[i,3]=lightness[i]
    data.iloc[i,4]=saturation[i]
    
    
    data.iloc[i,5]=hueeee[index][0]
    data.iloc[i,6]=hueeee[index][1]
    data.iloc[i,7]=hueeee[index][2]
    data.iloc[i,8]=hueeee[index][3]
    data.iloc[i,9]=hueeee[index][4]
    data.iloc[i,10]=hueeee[index][5]
    data.iloc[i,11]=hueeee[index][6]
    data.iloc[i,12]=hueeee[index][7]
    data.iloc[i,13]=hueeee[index][8]
    data.iloc[i,14]=hueeee[index][9]
    data.iloc[i,15]=hueeee[index][10]
    data.iloc[i,16]=hueeee[index][11]
    index = index +1
```
